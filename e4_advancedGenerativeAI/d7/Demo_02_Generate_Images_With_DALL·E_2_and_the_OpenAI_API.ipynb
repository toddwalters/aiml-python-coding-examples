# **Step 1: Setup the environment**

```python
import os
import openai
```

# **Step 2:  Generate image from the text**

```python
Text_PROMPT = input("Enter a Text : ")



Image_response = openai.Image.create(
    prompt=Text_PROMPT,
    n=1,
    size="256x256",
)

print(Image_response["data"][0]["url"])
```

# **Step 3: Display the generated images**
* Use the openai.Image.create() method to generate images based on a prompt
* Process the response to display the generated images


```python
import openai
import requests
from IPython.display import Image, display


# Generate images based on a prompt
response_data = openai.Image.create(
    prompt="A caricature of a cute happy puppy.",
    n=2,
    size="256x256"
)

# Display the generated images
if response_data and "data" in response_data:
    image_response = response_data["data"]
    for image in image_response:
        url = image.get("url")
        if url:
            display(Image(requests.get(url).content))
        else:
            print("Image URL not found in response.")
else:
    print("Image generation failed or no response received.")
```
